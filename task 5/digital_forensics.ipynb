```python
# Define a function to classify the equality score
def classify_score(Score):
    if abs(Score) <= 10:
        return 'Fair'
    elif abs(Score) > 20:
        return 'Highly Discriminative'
    else:
        return 'Unfair'
```

```python
# Add the Equality class column
data['Equality class'] = data['Equality Score'].apply(classify_score)
```
